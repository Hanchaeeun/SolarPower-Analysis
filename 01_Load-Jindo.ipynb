# Photovoltaics Data

```python
path = '/content/gdrive/MyDrive/SolarPower/Solardata/Jindo/'
Jd_df = Load_PV_Jindo(path)
Jd_df['지역'] = '전남 진도군'
Jd_df.head() 
```

```python
drop_cols =  ['발전기명', '발전기코드', '발전_펌핑구분', '충전_방전', '합계', '발전/펌핑구분']
Jd_df.drop(columns=drop_cols, inplace=True) 
Jd_df
```

```python
Jd_df.columns
```

```python
Jd_df = Format_PV(Jd_df, '날짜', ['날짜','지역','년도'])
Jd_df['일자'] = pd.to_datetime(Jd_df['날짜'])
Jd_df.drop(columns='날짜', inplace=True)
Jd_df
```

```python
Jd_df['년도'] = Jd_df['년도'].astype(int)
print('진도 : ', Jd_df.shape , Jd_df['년도'].unique())
```

```python
PV_Jd = Jd_df.sort_values(['년도','일자','시간'])
PV_Jd.reset_index(drop=True, inplace=True)
PV_Jd 
```

# Merge data

```python
print(PV_Jd.shape, ASOS_Jd.shape)
```

```python
print(ASOS_Jd['시간'].unique())
print(PV_Jd['시간'].unique())
```

```python
# 시간이 0-23이 24시를 0으로 변경 (밤의 데이터는 예측에 사용되지 않기에 임의의 숫자로 변경함)
PV_Jd['시간'] = PV_Jd['시간'].replace(24, 0)
```

```python
Jd_data = pd.merge(ASOS_Jd, PV_Jd, how='inner', on=['년도','일자','시간'])
Jd_data
```

# Rename

```python
cols_rename = {
    '일자' : 'Date',
    '지역' : 'Area',
    '시간' : 'Time',
    '발전량' : 'Photovoltaics',
    '강수량' : 'Precipitation',
    '기온' : 'Temperature',
    '풍속' : 'Wind velocity',
    '풍향' : 'Wind direction',
    '습도' : 'Humidity',
    '해면기압' : 'Sea-level pressure',
    '일조' : 'Sunshine',
    '일사' : 'Solar radiation',
    '전운량' : 'Cloudy',
    '시정' : 'Visibility',
    'PM10' : 'PM10',
    'PM25' : 'PM25',
    '월' : 'Month'
}
```

```python
Jd_data.rename(columns = cols_rename, inplace=True)
Jd_data.drop(columns=['년도','지점명'], inplace=True)
Jd_data['Month'] = pd.to_datetime(Jd_data['Date']).dt.month
Jd_data
```
