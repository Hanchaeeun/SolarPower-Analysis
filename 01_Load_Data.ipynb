## Particulate Matter Data

```python
path = '/content/gdrive/MyDrive/SolarPower/PMdata/'
PMdata = load_module.load_PM(path, '지역', ['경남 진주시', '강원 강릉시', '전남 목포시'])
PMdata
```

```python
for i in ['경남 진주시', '강원 강릉시', '전남 목포시']:
  data = PMdata[PMdata['지역'] == i ]
  area = data['측정소명'].unique()
  print(f'{i} : {data.shape}, 측정소 : {area}')
```

```python
Jinju_df = PMdata[PMdata['지역'] == '경남 진주시']
for i in ['상봉동','상대동(진주)','대안동', '상대동', '정촌면']:
  j = Jinju_df[Jinju_df['측정소명'] == i]
  print(f'{i} : {j.shape}')
```

```python
Mokpo_df = PMdata[PMdata['지역'] == '전남 목포시']
for i in ['용당동', '부흥동']:
  j = Mokpo_df[Mokpo_df['측정소명'] == i]
  print(f'{i} : {j.shape}')
```

```python
PMdata = PMdata[PMdata['측정소명'].isin(['대안동','옥천동','용당동'])]
PMdata = PMdata[['지역','일자','년도','시간','PM10','PM25']]
PMdata['지점명'] = PMdata['지역'].str[-3:-1]
PMdata.reset_index(drop=True, inplace=True)
PMdata
```

```python
# 측정 이상 데이터 제거
pm_idx = PMdata[PMdata['PM10'] == -999].index 
print(len(pm_idx))
PMdata.drop(pm_idx, axis=0, inplace=True)
```

## Photovoltaics Data

```python
path = '/content/gdrive/MyDrive/SolarPower/Solardata/Jinjo/'
Jj_df = load_module.load_PV(path)
Jj_df['지역'] = '경남 진주시'
Jj_df.head()
```

```python
path = '/content/gdrive/MyDrive/SolarPower/Solardata/Gangneung/'
Gn_df = load_module.load_PV(path)
Gn_df['지역'] = '강원 강릉시'
Gn_df.head()
```

```python
path = '/content/gdrive/MyDrive/SolarPower/Solardata/Yeongam/'
Ya_df = load_module.load_PV_Mokpo(path)
Ya_df['지역'] = '전남 목포시'
Ya_df.head()
```

```python
Ya_df['년월일'] = Ya_df['년도'].astype(str) + f'-' + Ya_df['월'].astype(str) + f'-' + Ya_df['일'].astype(str)
Ya_df.drop(columns=['월','일'],inplace=True)
Ya_df.head()
```

```python
print('강릉 : ',Gn_df.shape ,Gn_df['년도'].unique())
print('진주 : ',Jj_df.shape ,Jj_df['년도'].unique())
print('목포 : ',Ya_df.shape ,Ya_df['년도'].unique())
```

```python
PVdata = pd.concat([Jj_df, Gn_df, Ya_df], axis=0)
print(PVdata.shape)
PVdata.head(3)
```

```python
PVdata['일자'] = pd.to_datetime(PVdata['년월일'], format="%Y-%M-%d").dt.strftime("%Y-%M-%d")
PVdata.drop(columns='년월일', inplace=True)
PVdata.sort_values(['지역','년도','일자'], inplace=True)
PVdata.reset_index(drop=True, inplace=True)
PVdata
```
