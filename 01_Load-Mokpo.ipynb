## Photovoltaics Data

```python
path = '/content/gdrive/MyDrive/SolarPower/Solardata/Yeongam/'
Ya_PV = Load_PV_Mokpo(path)
Ya_PV['지역'] = '전남 목포시'
Ya_PV.head() 
```

```python
Ya_PV['년월일'] = Ya_PV['년도'].astype(str) + f'-' + Ya_PV['월'].astype(str) + f'-' + Ya_PV['일'].astype(str)
Ya_PV.drop(columns =['계', ' 계 ', '월', '일'], inplace=True)
Ya_PV.head()
```

```python
Ya_PV['일자'] = pd.to_datetime(Ya_PV['년월일'], format="%Y-%M-%d").dt.strftime("%Y-%M-%d")
Ya_PV.drop(columns='년월일', inplace=True)
Ya_PV.sort_values(['지역','년도','일자'], inplace=True)
Ya_PV.reset_index(drop=True, inplace=True)
Ya_PV
```

```python
PV_df = Format_PV(Ya_PV, '일자', ['일자','지역','년도'])
PV_df
```

```python
PV_df['년도'] = PV_df['년도'].astype(int)
print('목포 : ', PV_df.shape, PV_df['년도'].unique())
```

```python
PV_df = PV_df.sort_values(['년도','일자','시간'])
PV_df.reset_index(drop=True, inplace=True)
PV_df['일자'] = pd.to_datetime(PV_df['일자'])
PV_df 
```

# Merge data

```python
print(f'PV : {PV_df.shape}\nASOS : {Mokpo.shape}\nPM : {PMdata.shape}')
```

```python
print(Mokpo['시간'].unique())
print(PV_df['시간'].unique())
```

```python
# 시간이 0-23이 24시를 0으로 변경 (밤의 데이터는 예측에 사용되지 않기에 임의의 숫자로 변경함)
PV_df['시간'] = PV_df['시간'].replace(24, 0)
```

```python
Mp_data = pd.merge(Mokpo, PV_df, how='inner', on=['년도','일자','시간'])
Mp_data
```

# Rename

```python
cols_rename = {
    '일자' : 'Date',
    '지역' : 'Area',
    '시간' : 'Time',
    '발전량' : 'Photovoltaics',
    '강수량' : 'Precipitation',
    '기온' : 'Temperature',
    '풍속' : 'Wind velocity',
    '풍향' : 'Wind direction',
    '습도' : 'Humidity',
    '해면기압' : 'Sea-level pressure',
    '일조' : 'Sunshine',
    '일사' : 'Solar radiation',
    '전운량' : 'Cloudy',
    '시정' : 'Visibility',
    'PM10' : 'PM10',
    'PM25' : 'PM25',
    '월' : 'Month'
}
```

```python
Mp_data.rename(columns = cols_rename, inplace=True)
Mp_data.drop(columns=['년도','지점명'], inplace=True)
Mp_data['Month'] = pd.to_datetime(Mp_data['Date']).dt.month
Mp_data
```
